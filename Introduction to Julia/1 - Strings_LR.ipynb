## How to get a string

Enclose your characters in " " or """ """!

````julia
s1 = "I am a string."
````

````julia
s2 = """I am also a string. """
````

There are a couple functional differences between strings enclosed in single and triple quotes. <br>
One difference is that, in the latter case, you can use quotation marks within your string.

````julia
"Here, we get an "error" because it's ambiguous where this string ends "
````

````julia
#use this if we want to use quotation marks within the string
"""Look, Mom, no "errors"!!! """
````

Note that ' ' define a character, but NOT a string!

````julia
#this is how we create a character variable but NOT a string
typeof('a')
````

````julia
#this is not a character but a proper sentence/string variable we would like to declare
'We will get an error here'
````

## String interpolation

We can use the $ sign to insert existing variables into a string and to evaluate expressions within a string. <br>
Below is an example that contains some highly sensitive personal information.

````julia
name = "Jane"
num_fingers = 10
num_toes = 10
````

````julia
#we can use the dollar sign to replace with the actual variable, e.g. name, num_fingers and num_toes
println("Hello, my name is $name.")
println("I have $num_fingers fingers and $num_toes toes.")
````

````julia
# we can also use this to perform calculations within a string! 
println("That is $(num_fingers + num_toes) digits in all!!")
````

## String concatenation

Below are three ways we can concatenate strings! <br>
#### 1) Using the string function to connect strings
The first way is to use the `string()` function. <br>
`string()` converts non-string inputs to strings.

````julia
# we can use string to connet two string.
string("How many cats ", "are too many cats?")
````

````julia
# we can even add integers in between the strings
string("I do not know, but ", 10, " are too few.")
````

#### 2) Storing strings in variables and using string()

````julia
# we can store the strings stored in variables
s3 = "How many cats ";
s4 = "is too many cats?";
ðŸ˜º = 10
````

````julia
string(s3, s4)
````

````julia
string("I don't know, but ", ðŸ˜º, " is too few.")
````

#### 3) Storing strings in variables and using *
We can also use `*` for concatenation!

````julia
s3*s4
````
